```python
Question 11
```

```python
df_encoded = df.copy()

label_encoders = {}

categorical_cols = [
    "Gender",
    "SubscriptionType",
    "FavoriteGenre",
    "SubscriptionRenewed"
]

for col in categorical_cols:
    le = LabelEncoder()
    df_encoded[col] = le.fit_transform(df_encoded[col])
    label_encoders[col] = le

print("\nEncoded Dataset:")
print(df_encoded.head())
```

```python
Question 12
```

```python
X = df_encoded.drop("SubscriptionRenewed", axis=1)
y = df_encoded["SubscriptionRenewed"]

print("\nFeature Matrix Shape:", X.shape)
print("Target Shape:", y.shape)
```

```python
Question 13
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42
)

print("\nTraining Samples:", len(X_train))
print("Testing Samples:", len(X_test))
```

```python
Question 14
```

```python
dt_model = DecisionTreeClassifier(
    random_state=42
)

dt_model.fit(X_train, y_train)
```

```python
Question 15
```

```python
dt_pred = dt_model.predict(X_test)

dt_accuracy = accuracy_score(
    y_test,
    dt_pred
)

print("\nDecision Tree Accuracy:",
      round(dt_accuracy * 100, 2), "%")
```

```python
Question 16
```

```python
cm = confusion_matrix(
    y_test,
    dt_pred
)

print("\nConfusion Matrix:")
print(cm)
```

```python
Question 17
```

```python
knn_model = KNeighborsClassifier(
    n_neighbors=5
)

knn_model.fit(X_train, y_train)
```

```python
Question 18
```

```python
knn_pred = knn_model.predict(X_test)

knn_accuracy = accuracy_score(
    y_test,
    knn_pred
)

print("\nKNN Accuracy:",
      round(knn_accuracy * 100, 2), "%")

print("\nModel Comparison")
print("------------------")
print("Decision Tree:", round(dt_accuracy*100,2), "%")
print("KNN          :", round(knn_accuracy*100,2), "%")

if knn_accuracy > dt_accuracy:
    print("KNN performed better.")
else:
    print("Decision Tree performed better.")
```

```python
Q2. Why is predicting subscription renewal considered a classification problem?

Predicting subscription renewal is a classification problem because the target variable has two possible categories: Yes (renewal) or No (non-renewal). The model assigns each customer to one of these classes based on their features.
```

```python
Q3. Why is predicting monthly spending considered a regression problem?

Predicting monthly spending is a regression problem because the target variable is a continuous numerical value. The model estimates the amount of money a customer is likely to spend rather than assigning a category.
```
